```python
import numpy as np
import pandas as pd

print("\n\n-------------------------------------------------------")
print("                       QUESTION 1")
print("-------------------------------------------------------")

A = np.array([1, 2, 3, 2, 3, 4, 3, 4, 5, 6])
B = np.array([7, 3, 3, 2, 10, 2, 7, 4, 9, 4, 9, 8])

C = []
for i in range(0,min(len(A),len(B)), 1):
    if A[i] == B[i]:
        C.append(i)
print(C)


print("\n\n-------------------------------------------------------")
print("                       QUESTION 2")
print("-------------------------------------------------------")

fileData = pd.read_excel("automobile_data.xlsx")

#PART 1
print("\nPrint first five entries of data set : ")
print(fileData.head(6))  # as first is a title of each COLUMN

#PART 2
print("\nPrint last five entries of data set : ")
print(fileData.tail())

#PART 3
print("\nPrint columns of data set :")
print(fileData.columns)


#PART 4 + 5
print("\nPrint unique values + NULL count for each column :")
for data in fileData:
    print("\nUnique value count in column ", end=""),
    print(data, end=" : ")
    print(fileData[data].nunique())
    print("Sum of NULL Values in column ", end=""),
    print(data, end=" : ")
    print(fileData[data].isnull().sum())

#PART 6
print("\nRemove unnecessary column :")
new_data = fileData.drop(['index'], axis=1)
print(fileData)

#PART 7
print("\nDrop NULL values :")
fileData.dropna(subset=fileData.columns, inplace=True)
print(fileData)

print("\n\n-------------------------------------------------------")
print("                       QUESTION 3")
print("-------------------------------------------------------")
fileData = pd.read_csv("automobile_data.csv")

#PART 1
print("\nMost expensive car from the dataset")
car = fileData[fileData['price'] == max(fileData['price'])].to_string(index=True).strip()
print(car)


#PART 2
print("\nCalculate total cars per manufacturer and show the result")
Companies = fileData['company'].unique()
for i in Companies:
    print(i, end=" ")
    print(len(fileData[fileData['company'] == i]))


#PART 3
print("\nThe details of vehicles against Toyota manufacturer ")
groupByCompany = fileData.groupby('company')
print(groupByCompany.get_group('toyota'))

#PART 4
print("\nArrange the cars according to the prices (highest-lowest) and display relevant information for first 5 rows  ")
sortData = fileData.sort_values(by=['price'], ascending=False)
print(sortData.head(5))
```
